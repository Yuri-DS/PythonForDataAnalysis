#### Часть 3: Операции над массивами

#### Арифметические операции:

- Создайте два одномерных массива из 10 случайных чисел и выполните поэлементное сложение, вычитание, умножение и деление этих массивов.
- Найдите скалярное произведение двух одномерных массивов.

#### Матричные операции:

- Создайте две матрицы размером 3x3 и выполните их матричное умножение.
- Найдите обратную матрицу для одной из созданных матриц.



```python
arr1 = np.random.randint(1, 50, size=10)
arr2 = np.random.randint(50, 101, size=10)

print(arr1, arr2, "\n")

print(f"сложение - {arr2 + arr1}")
print(f"вычитание - {arr2 - arr1}")
print(f"умножение - {arr2 * arr1}")
print(f"деление - {arr2 / arr1}")
```

```python
arr1 = np.random.randint(1, 10, size=10)  
arr2 = np.random.randint(1, 10, size=10) 

print(arr1, arr2)

print("Скалярное произведение:", np.dot(arr1, arr2))
```

```python
A = np.random.randint(1, 10, size=(3, 3))
B = np.random.randint(1, 10, size=(3, 3))

print("Первая матрица A:\n", A)
print("Вторая матрица B:\n", B)

print("Результат матричного умножения C = A @ B:\n", np.dot(A, B))
```

```python
A_inv = np.linalg.inv(A)

print("Обратная матрица для A:\n", A_inv)
```


### Часть 4: Анализ данных

- Создание массива:

Создайте двумерный массив размером 5x3, где первый столбец содержит имена клиентов, второй — их возраст, а третий — стоимость ремонта их автомобилей.


- Сортировка массива:

Отсортируйте массив по возрасту клиентов.

- Средние значения:

Найдите средний возраст клиентов и среднюю стоимость ремонта.

Максимальная стоимость ремонта:

Найдите клиента с максимальной стоимостью ремонта.

```python
data = np.array([
    ('Иван', 25, 15000),
    ('Анна', 30, 25000),
    ('Петр', 22, 18000),
    ('Мария', 35, 30000),
    ('Сергей', 28, 22000)
], dtype=object)

# Получаем индексы для сортировки по возрасту (второй столбец)
sorted_indices = np.argsort(data[:, 1])

# Сортируем массив по индексам
sorted_data = data[sorted_indices]

print("Отсортированный массив по возрасту:\n", sorted_data)

print(f"\nСредний возраст клиентов - {data[:, 1].mean()} \nСредняя стоимость ремнота - {data[:, 2].mean()}\n")

# Находим индекс клиента с максимальной стоимостью ремонта
max_cost_index = np.argmax(data[:, 2])

print(f"Максимальная стоимость ремонта {data[max_cost_index, 2]}\nКлиент с максимальной стоимостью ремонта - {data[max_cost_index, 0]}")
```
