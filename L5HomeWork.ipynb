### Домашнее задание
- из информационной системы московской биржи получить данные истории 2 ценных бумаг за год (можно больше). Например Сбербанк (SBER) и Московская биржа (MOEX). Анализируемы бумаги можно выбрать самостоятельно. Источник получения данных любой.
- Оставить данные только основной площадки BOARDID=TQBR
- Сделать TRADEDATE индексом, индекс должен быть типа datetime
- Сохранить данные в формате HDF5
- Посчитать корреляцию и ковариацию средневзвешенных цен (WAPRICE)

Пример url для получения данных:

https://iss.moex.com/iss/history/engines/stock/markets/shares/securities/SBER.csv?from=2023-01-01

https://iss.moex.com/iss/history/engines/stock/markets/shares/securities/MOEX.csv?from=2023-01-01

где SBER - тикер ценной бумаги

.csv - формат данных

from=2024-06-24 - дата начала получения исторических данных

start=0 - смещение

```python
import pandas as pd
from datetime import datetime


def date_parser(date):
    return pd.to_datetime(date, format='%Y-%m-%d')

url = "https://iss.moex.com/iss/history/engines/stock/markets/shares/securities/SBER.csv?from=2023-01-01"
df_sber= pd.read_csv(url, encoding='cp1251', skiprows=2, delimiter=';', engine="python")

url = "https://iss.moex.com/iss/history/engines/stock/markets/shares/securities/MOEX.csv?from=2023-01-01"
df_moex = pd.read_csv(url, encoding='cp1251', skiprows=2, delimiter=';', engine="python")

# df_sber.head()
# df_moex.head()
```

```python
# Оставить данные только основной площадки BOARDID=TQBR
df_sber = df_sber[df_sber["BOARDID"] == "TQBR"]

df_moex = df_moex[df_moex["BOARDID"] == "TQBR"]

# df_sber.head()
# df_moex.head()
```

```python
# Сделать TRADEDATE индексом, индекс должен быть типа datetime
df_sber['TRADEDATE'] = pd.to_datetime(df_sber['TRADEDATE'])

df_sber.set_index('TRADEDATE', inplace=True)

df_sber.head()
```

```python
# Сделать TRADEDATE индексом, индекс должен быть типа datetime
df_moex['TRADEDATE'] = pd.to_datetime(df_moex['TRADEDATE'])

df_moex.set_index('TRADEDATE', inplace=True)

df_moex.head()
```

```python
# Сохранить данные в формате HDF5
df_sber.to_hdf('data_sber.h5', key='moex_data', mode='w')
df_moex.to_hdf('data_moex.h5', key='moex_data', mode='w')
```

```python
# Посчитать корреляцию и ковариацию средневзвешенных цен (WAPRICE)
print(df_sber["WAPRICE"].cov(df_moex["WAPRICE"]))   # ковариация 
print(df_sber["WAPRICE"].corr(df_moex["WAPRICE"]))  # корреляция
```
